#Bashの環境変数<br>
環境変数は、システムやアプリケーションの動作を制御するために使われる変数です。Bashの環境変数は、主に以下のようなものがあります。

**PATH**: 実行ファイルの検索パスを定義します。ディレクトリのリストをコロンで区切って指定します。<br>
**HOME**: ユーザーのホームディレクトリを示します。<br>
**USER**: 現在のユーザー名を示します。<br>
**SHELL**: 現在のシェルのパスを示します。<br>
**PS1**: プロンプトの外観を設定します。<br>
環境変数を表示するにはprintenvやenvコマンドを使います。設定や変更は以下のように行います。


```
export VARIABLE_NAME=value
```


#.bash_profile, .bashrc, .bashenv
これらのファイルは、シェルの初期化時に読み込まれるスクリプトです。

**.bash_profile**: ログインシェルが起動する際に実行されます。一般的に、環境変数の設定やパスの設定を行います。.bash_loginやprofileも同様の役割を果たしますが、優先順位が異なります。

**.bashrc**: 非ログインシェル（例えば、新しいターミナルウィンドウやタブを開いたとき）で実行されます。エイリアスやシェルの設定が多く含まれます。

**.bashenv**: すべてのシェル（ログインシェル、非ログインシェル）で実行されます。一般的には環境変数の設定に使われますが、利用頻度は低いです。

#fg, bg, ps, kill コマンド
**fg**: バックグラウンドで実行中のジョブをフォアグラウンドに戻します。


```
fg %1
```


**bg**: フォアグラウンドで一時停止したジョブをバックグラウンドで実行し続けます。


```
bg %1
```


**ps**: 現在のプロセスの情報を表示します。例えば、すべてのプロセスを表示するにはps -eを使用します。


```
ps -aux
```


**kill**: プロセスにシグナルを送ります。一般的にプロセスを終了するために使われます。


```
kill 1234
```


#パイプ、リダイレクト
**パイプ（|）**: 一つのコマンドの出力を別のコマンドの入力として渡します。


```
ls -l | grep "txt"
```


**リダイレクト（>、>>、<）**: コマンドの出力をファイルに保存したり、ファイルの内容をコマンドの入力として読み込んだりします。


```
ls > file.txt  # 出力をfile.txtに上書き保存
ls >> file.txt # 出力をfile.txtに追記
cat < file.txt # file.txtの内容をcatコマンドの入力として読み込む
```


#grep, less コマンド
**grep**: ファイル内の文字列を検索します。


```
grep "search_string" file.txt
```


**less**: ファイルの内容をページ単位で表示します。moreコマンドに似ていますが、前後にスクロールできます。


```
less file.txt
```
